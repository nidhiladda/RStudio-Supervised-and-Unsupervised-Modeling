```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
from google.colab import drive
#drive.mount("/content/drive")
df = pd.read_csv("/content/drive/MyDrive/CASchools.csv")
```

```python
# Basic Data Analysis
def analyze_data(df):
    print("Data Summary:")
    print(df.describe())
    print("\nMissing Values:")
    print(df.isnull().sum())

# Visualize Data
def plot_data(df):
    plt.figure(figsize=(14, 8))

    # Plot average reading scores
    plt.subplot(2, 2, 1)
    sns.histplot(df['read'], bins=20, kde=True)
    plt.title('Average Reading Scores')

    # Plot average math scores
    plt.subplot(2, 2, 2)
    sns.histplot(df['math'], bins=20, kde=True)
    plt.title('Average Math Scores')

    # Plot student enrollment distribution
    plt.subplot(2, 2, 3)
    sns.histplot(df['students'], bins=20, kde=True)
    plt.title('Student Enrollment Distribution')

    # Plot expenditure per student
    plt.subplot(2, 2, 4)
    sns.histplot(df['expenditure'], bins=20, kde=True)
    plt.title('Expenditure per Student')

    plt.tight_layout()
    plt.show()

# Main function
def main():
    # Analyze the data
    analyze_data(df)

    # Plot the data
    plot_data(df)

if __name__ == "__main__":
    main()
```
